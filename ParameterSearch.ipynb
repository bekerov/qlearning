```python
%matplotlib inline

from agents import trainer

trainer = trainer.Trainer('CartPole-v0')

explored_parameters = {}
for i in range(10):
    parameters = trainer.gen_parameters()
    results = []
    print str(i), str(parameters)
    for _ in range(10):
        results.append(trainer.train(target_reward=100,
                                          max_epochs=5000, parameters=parameters, out=True))
    explored_parameters[parameters] = results
print "Graphing"
for parameters, results in explored_parameters.iteritems():
    import matplotlib.pyplot as plt

    print str(parameters)
    plt.hist([r['reward'] for r in results])
    plt.title(str(parameters))
    plt.show()
```
